```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import numpy as np
import pickle
```

```python
data = pd.read_csv("dataset/data.csv")
data.head(5)
```

```python
data.info()
```

```python
temp = data.iloc[:, 3:]
temp.head()
```

```python
data.columns
```

```python
data.drop(data.columns[0:3], axis=1, inplace=True)
data.head()
```

```python
data.columns
```

```python
newColumns = ["Method", "Mechanism", "Exam", "Interaction", "Contact", "Efficiency"]
newColumns
```

```python
for i in range(6):
    data.rename(columns={ data.columns[i] : newColumns[i] }, inplace=True)
data.head()
```

```python
#Label Encoder
labelEncoder = LabelEncoder()
labelEncoder
```

```python
data["Method"] = labelEncoder.fit_transform(data["Method"])
data.head()
```

```python
for column in data.columns[1:]:
    data[column] = labelEncoder.fit_transform(data[column])
data.head()
```

```python
#One Hot Encoder
oneHotEncoder = OneHotEncoder()
oneHotEncoder
```

```python
method = oneHotEncoder.fit_transform(data.iloc[:, 0:1]).toarray().astype(int)
type(method)
```

```python
method = pd.DataFrame(data = method, index = range(969), columns = ["Sync", "Async", "Both"])
method.head()
```

```python
data = pd.concat([method, data.iloc[:, 1:]], axis=1)
data.head()
```

```python
#Train Test Split
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,0:-1], data.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
X_train
```

```python
# Linear Regression
regressor = LinearRegression()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
# Linear Regression Predict
y_pred = regressor.predict(X_test)
y_pred
```
