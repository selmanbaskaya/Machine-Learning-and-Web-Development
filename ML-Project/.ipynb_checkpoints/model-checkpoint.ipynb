```python
#Import Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#Dataset Upload
data = pd.read_csv("dataset/data.csv")
data.head(5)
```

```python
data.info()
```

```python
temp = data.iloc[:, 3:]
temp.head()
```

```python
data.columns
```

```python
#Column Drop in Dataset
data.drop(data.columns[0:3], axis=1, inplace=True)
data.head()
```

```python
#Changing Column Names
newColumns = ["Method", "Mechanism", "Exam", "Interaction", "Contact", "Efficiency"]
newColumns
```

```python
for i in range(6):
    data.rename(columns={ data.columns[i] : newColumns[i] }, inplace=True)
data.head()
```

```python
#Label Encoder
labelEncoder = LabelEncoder()
labelEncoder
```

```python
#Single Label Encoder
data["Method"] = labelEncoder.fit_transform(data["Method"])
data.head()
```

```python
#Multiple Label Encoder
for column in data.columns[1:]:
    data[column] = labelEncoder.fit_transform(data[column])
data.head()
```

```python
#One Hot Encoder
oneHotEncoder = OneHotEncoder()
oneHotEncoder
```

```python
method = oneHotEncoder.fit_transform(data.iloc[:, 0:1]).toarray().astype(int)
type(method)
```

```python
#Converting to Pandas DataFrame
method = pd.DataFrame(data = method, index = range(969), columns = ["Sync", "Async", "Both"])
method.head()
```

```python
#Concating
data = pd.concat([method, data.iloc[:, 1:]], axis=1)
data.head()
```
